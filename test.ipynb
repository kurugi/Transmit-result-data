```python
# 병원결과.xlsx 3번째 행 해더 읽어오기기
import openpyxl

def print_third_row(filename):
    # 엑셀 파일 열기
    wb = openpyxl.load_workbook(filename, data_only=True)
    ws = wb.active  # 기본 활성화된 시트 선택
    
    # 3번째 행 가져오기
    third_row = [cell.value for cell in ws[3]]
    
    # 출력
    print("3번째 줄 데이터:", third_row)

    # 파일 닫기
    wb.close()

# 사용 예시
print_third_row("병원결과.xlsx")
```

```python
import openpyxl

# 컬럼 매핑 (병원결과 → LG결과)
column_map = {
    "사원번호": "EMP_NO",
    "주민등록번호": "SSN",
    "진료일자": "MDC_DATE",    
    "HA001": "BM01",
    "HA002": "BM02",
    "HA004": "BM04",
    "FAT": "BM05",
    "HA007": "BM07",
    "L90001": "BT01",
    "L70013": "BT02",
    "HB001": "CV02",
    "HB002": "CV01",
    "A8001": "CV03",
    "E6541": "CV04",
    "HO001": "OE101",
    "HO002": "OE102",
    "HO003": "OE103",
    "HO004": "OE104",
    "HO011": "OE205",
    "HO005": "OE301",
    "HO006": "OE302",
    "OHA01": "AM103",
    "OHA09": "AM104",
    "OHA02": "AM105",
    "OHA10": "AM106",
    "OHA03": "AM107",
    "OHA11": "AM108",
    "OHA04": "AM121",
    "OHA12": "AM122",
    "OHA05": "AM111",
    "OHA13": "AM112",
    "OHA06": "AM115",
    "OHA14": "AM116",
    "L2012": "CB101",
    "L2014": "CB110",
    "L07003501": "CB112",
    "L20220": "CB113",
    "L20141": "CB104",
    "L20142": "CB105",
    "L20143": "CB106",
    "L2015": "CB107",
    "L2016": "CB108",
    "L2011": "CB109",
    "L20191": "CB203",
    "L20193": "CB204",
    "L20192": "CB205",
    "L20194": "CB206",
    "L3012": "DM04",
    "L3231": "DM07",
    "L3015": "LP01",
    "L3081": "LP02",
    "L3082": "LP03",
    "L3083": "LP04",
    "L3016": "LF13",
    "LAC10401": "LF14",
    "A009": "LF15",
    "L3018": "LF03",
    "L3019": "LF04",
    "L3033": "LF05",    
    "L3020": "LF06",
    "L3062": "LF11",
    "LAC10402": "LF10",
    "L3051": "LF16",
    "LAC114": "LF17",
    "LAC158": "LF19",
    "N20501": "VE105",
    "N20502": "VE102",
    "N20511": "VE201",
    "L170360": "VE106",
    "L170380": "VE107",
    "L3032": "RF03",
    "L3013": "RF02",
    "LAC104031": "RF04",
    "LAC13701": "EL01",
    "L3031": "EL02",
    "L3041": "EL03",
    "L3042": "EL04",
    "RU103": "TF01",
    "N20006": "TF06",
    "N20003": "TF03",
    "LIS11001": "VE301",
    "L3092": "SY03",
    "L51821": "CE01",
    "L51811": "CE02",
    "N206101": "CE03",
    "N20609": "CE04",
    "N206151": "CE05",
    "L150026": "CE06",
    "L3014": "GT01",
    "L5114": "RA03",
    "L6103": "UA101",
    "L6102": "UA102",
    "L6110": "UA103",
    "L6104": "UA106",
    "L6107": "UA108",
    "L6108": "UA301",
    "L6109": "UA302",
    "L6105": "UA401",
    "L61132": "UA201",
    "L61131": "UA202",
    "L61133": "UA203",
    "TK531": "SE02",
    "TK02": "SE01",
    "TK03": "SE03",
    "RZ901A2": "BD01",
    "RP201": "RE101",
    "L9742HPC": "RE200",
    "RC4011": "GI303",
    "S1005": "GI201",
    "L5237": "GI203",
    "C5602": "GI205",
    "RU401": "US02",
    "S2322": "US03",
    "RU902A": "US04",
    "RU505": "US05",
    "S1008B": "US07",
    "CTN710": "US08",
    "N456232": "US09",
    "RC101": "US10",
    "C5601": "US13",
    # "MC090": "GY03",
    "RP20BBA": "GY03",
    "RU201": "GY04",
    "P30001": "GY05",
    "RC94HL": "RE402",
    "L2013": "CB102",
    "L3021": "LF12",
    "L1820": "LF08",
    "LAC162": "RA01",
    "TH01": "RA02",
    "L20221": "CB111",
    "L20190": "CB201",
    "RC94HC": "RE403",
    "L6106": "UA402",
    "L5237":"G1203",
    "RN801":"US15",
    "N456000":"US16",
    "N455004":"US17",
    "CTN711H":"US20",
    "SE60":"US11",
    "TX0B300":"GY07"
}

def clear_existing_data_from_row(ws, start_row=4):
    """
    LG결과.xlsx의 start_row 행부터 모든 데이터를 삭제
    """
    max_row = ws.max_row
    if max_row >= start_row:
        for _ in range(start_row, max_row + 1):
            ws.delete_rows(start_row)  # start_row 행을 계속 삭제하면 자동으로 위로 당겨짐

def transform_hospital_results(input_file, output_file):
    """
    병원결과.xlsx → LG결과.xlsx 변환 및 데이터 업데이트 (3행까지 유지)
    """
    
    # 병원결과.xlsx 열기
    wb_input = openpyxl.load_workbook(input_file, data_only=True)
    ws_input = wb_input.active

    # LG결과.xlsx 열기 (없으면 생성)
    try:
        wb_output = openpyxl.load_workbook(output_file)
        ws_output = wb_output.active
    except FileNotFoundError:
        wb_output = openpyxl.Workbook()
        ws_output = wb_output.active

    # LG결과.xlsx에서 4행부터 데이터 삭제
    clear_existing_data_from_row(ws_output, start_row=4)

    # 병원결과.xlsx에서 원본 헤더 읽기
    hospital_headers = [cell.value for cell in ws_input[1]]

    # LG결과.xlsx에서 기존 헤더 읽기 (1~3행 중 마지막 행을 기준으로 헤더 설정)
    lg_headers = [cell.value for cell in ws_output[3]]  # LG결과.xlsx의 3행을 헤더로 사용

    # 병원결과.xlsx 데이터 변환 및 삽입
    for row in ws_input.iter_rows(min_row=2, values_only=True):  # 데이터만 읽기
        mapped_row = {column_map.get(h, h): v for h, v in zip(hospital_headers, row)}  # 병원 헤더 → LG 헤더 변환
        transformed_row = [mapped_row.get(header, None) for header in lg_headers]  # LG결과 헤더에 맞게 정렬
        ws_output.append(transformed_row)  # 4행부터 삽입

    # LG결과.xlsx 저장
    wb_output.save(output_file)
    print(f"변환 완료: {output_file}")

# 실행 예시
transform_hospital_results("병원결과.xlsx", "LG결과.xlsx")
```
